### 2. Analysis of data of first four months (January, February, March and April) of last four years including 2020

**Since I will be taking 'Vehicular and Industrial Pollution' as key point hence my I will be focusing on Sulfer dioxide ($So_2$) and Nirtogen dioxide ($No_2$)**<br>
**As we know the main source for $So_2$ are power plants and it comes from burning of coal and oil.**<br>
**And the main source for $No_2$ are vehicels and it comes from burning fossil fuels (like gasoline)**

**I am taking few Industrial cities like Delhi, Bengaluru, Gurugram etc. for observation of Pollution**

```python
city = pd.read_csv('city_day.csv')
# city.head()
```

```python
city_days =city.drop(['AQI_Bucket','Benzene','Toluene','Xylene','NOx','NO','CO','NH3','O3'],axis=1)
```

```python
city_days.dropna(inplace=True)
```

```python
city_days['Date'] = pd.to_datetime(city_days['Date'])
```

```python
city_days['year']= city_days.Date.dt.year
```

```python
city_days['month'] = city_days.Date.dt.month
```

```python
four = city_days[city_days.month<=4]
```

```python
four=four.dropna()
```

```python
grouped1 = four.groupby(['City','year','month']).mean().reset_index()
```

```python
grouped2 =four.groupby(['City','year','Date']).mean().reset_index()
```

```python

output = wt.Output()
plot_output = wt.Output()
aqi_output = wt.Output()
so2_output = wt.Output()
no2_output = wt.Output()


city =wt.Dropdown(
    options=['Amaravati', 'Bengaluru',
    'Delhi',
       'Gurugram', 'Hyderabad', 'Jaipur',
       'Kolkata', 'Mumbai',
        'Thiruvananthapuram'],
    value='Delhi',
    description='City:',
    disabled=False,
)

year =wt.Dropdown(
    options=['All',2015,2016,2017,2018,2019,2020],
    value='All',
    description='Year:',
    disabled=False,
)


def commonFiltering(city,year):
    output.clear_output()
    plot_output.clear_output()
    aqi_output.clear_output()
    so2_output.clear_output()
    no2_output.clear_output()
    
    if year == 'All':
        filters1 = grouped1[grouped1.City== city]
        filters2 = grouped2[grouped2.City==city]
    else:
        filters1 = grouped1[(grouped1.year==year)&(grouped1.City== city)]
        filters2 = grouped2[(grouped2.year==year) &(grouped2.City==city)]

    with output:
        display(filters1)
        
    with plot_output:
        if((year =='All')):
            fig = go.Figure(data=[
                        go.Bar(name='Pm2.5', x=filters1.month, y=filters1['PM2.5'],text=filters1.year,textposition='inside'),
                        go.Bar(name='PM10', x=filters1.month, y=filters1.PM10,text=filters1.year,textposition='inside'),
                        go.Bar(name='AQI', x=filters1.month, y=filters1.AQI,text=filters1.year,textposition='inside')
                        ],
                   layout = go.Layout(barmode='group',title=filters1.City.unique()[0],xaxis={'title':'Month'}, yaxis={'title':'Value'}))
  
            fig.show()
        elif year in list(filters1.year.unique()):
            fig = go.Figure(data=[
                        go.Scatter(name='Pm2.5', x=filters2.Date, y=filters2['PM2.5']),
                        go.Scatter(name='PM10', x=filters2.Date, y=filters2.PM10),
                        go.Scatter(name='AQI', x=filters2.Date, y=filters2.AQI)
                        ],
                   layout = go.Layout(barmode='group',title=filters1.City.unique()[0],xaxis={'title':'Month'}, yaxis={'title':'Value'}))
  
            fig.show()
        else:
            print('Sorry! The data for this year is not available')
            
    with no2_output:
        if year == 'All':
             fig = go.Figure(data=[
                        go.Bar(name='NO2', x=filters1.month, y=filters1['NO2'],text=filters1.year,textposition='inside'),
                        ],
                   layout = go.Layout(barmode='group',xaxis={'title':'Month'}, yaxis={'title':'NO2'}))
  
             fig.show()
            
        elif((year in list(filters2.year.unique()))):
            fig = go.Figure([go.Scatter(x=filters2['Date'],y=filters2['NO2'], name='NO2'),
                        ],
                        layout=go.Layout(xaxis={'title':'Date'}, yaxis={'title':'No2'})
                       )
            fig.update_layout(autosize = False,
                             width=500,
                              height=500
                             )
            fig.show()
        else:
            print('Sorry! The data for this year is not available')
    with so2_output:
        if year == "All":
            fig = go.Figure(data=[
                        go.Bar(name='SO2', x=filters1.month, y=filters1['SO2'],text=filters1.year,textposition='inside'),
                        ],
                   layout = go.Layout(barmode='group',xaxis={'title':'Month'}, yaxis={'title':'SO2'}))
  
            fig.show()
        elif((year in list(filters2.year.unique()))):
            fig = go.Figure([go.Scatter(x=filters2['Date'],y=filters2['SO2'], name='SO2'),
                            
                        ],
                        layout=go.Layout(title=filters2.City.unique()[0],xaxis={'title':'Date'}, yaxis={'title':'So2'})
                       )
            fig.update_layout(autosize = False,
                             width=500,
                              height=500
                             )
            fig.show()
        else:
            print('Sorry! The data for this year is not available')
        
        
        
    
def dropdown_city(change):
    commonFiltering(change.new,year.value)
    
def dropdown_year(change):
    commonFiltering(city.value,change.new)

city.observe(dropdown_city, names='value')
year.observe(dropdown_year, names='value')
```

```python
layout_out = wt.Layout(margin = '0 0 10px 0', width='100%')
input_box = wt.HBox([city,year],layout=layout_out)
```

```python
tab = wt.VBox([output, plot_output],layout=layout_out)
```

```python
box = wt.HBox([no2_output,so2_output],layout=layout_out)
```

#### 2.1 Comparision of  average Air Quality Index of each month of 2020 with respective months of previous years

```python
dashboard = wt.VBox([input_box,tab,box],layout=layout_out)

display(dashboard)
```

**As we can observe average Air Quality Index is slighlity less that months of its previous years**<br>
**However if we observe carefull we can see that the the air quality index is decreasing rapidly in 2020 (specially after lockdown)**<br>
**And we can also see decrease in $No_2$ and $So_2$ after lockdown in India.**


```python
rest_days = city_days[city_days.year !=2020]
```

```python
output1 = wt.Output()
so2_output1 = wt.Output()
no2_output1 = wt.Output()


city1 =wt.Dropdown(
    options=['Amaravati', 'Bengaluru',
    'Delhi',
       'Gurugram', 'Hyderabad', 'Jaipur',
       'Kolkata', 'Mumbai',
        'Thiruvananthapuram'],
    value='Delhi',
    description='City:',
    disabled=False,
)

year1 =wt.Dropdown(
    options=[2015,2016,2017,2018,2019],
    value=2019,
    description='Year:',
    disabled=False,
)


def commonFiltering1(city1,year1):
    output1.clear_output()
    so2_output1.clear_output()
    no2_output1.clear_output()
    filters = rest_days[(rest_days.year==year1)&(rest_days.City== city1)]
        

    with output1:
        display(filters)
    
    with no2_output1:
        if((year1 in list(filters.year.unique()))):
            fig = go.Figure([go.Scatter(x=filters['Date'],y=filters['NO2'], name='NO2'),
                        ],
                        layout=go.Layout(title=filters.City.unique()[0],xaxis={'title':'Date'}, yaxis={'title':'No2'})
                       )
            fig.update_layout(autosize = False,
                             width=500,
                              height=500
                             )
            fig.show()
        else:
            print('Sorry! The data for this year is not available')
    with so2_output1:
        if((year1 in list(filters.year.unique()))):
            fig = go.Figure([go.Scatter(x=filters['Date'],y=filters['SO2'], name='SO2'),
                        ],
                        layout=go.Layout(title=filters.City.unique()[0],xaxis={'title':'Date'}, yaxis={'title':'So2'})
                       )
            fig.update_layout(autosize = False,
                             width=500,
                              height=500
                             )
            fig.show()
        else:
            print('Sorry! The data for this year is not available')
    

def dropdown_city1(change):
    commonFiltering1(change.new,year1.value)
    
def dropdown_year1(change):
    commonFiltering1(city1.value,change.new)

city1.observe(dropdown_city1, names='value')
year1.observe(dropdown_year1, names='value')

        
```

```python
input_box1 = wt.HBox([city1,year1],layout=layout_out)
```

```python
box1 = wt.HBox([no2_output1,so2_output1],layout=layout_out)
```

```python
dashboard1 = wt.VBox([input_box1,box1],layout=layout_out)

display(dashboard1)
```

**If we observe the pattern in data of previous years then we can see that whenever there is decrease in the $NO_2$ and $So_2$ there is high increase after that in quanitity of $So_2$ and $No_2$.**
